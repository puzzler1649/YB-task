#### df_x에서 df_y를 left join시키기 (quality를 기준으로)

```python
left_join=pd.merge(left=df_x, right=df_y, how="left", on="quality")
left_join
```

#### birthDate 컬럼으로 age 컬럼 추가하기

```python
data.birthDate = data.birthDate.astype('datetime64')
data.head()
```

##### 2) age 컬럼 생성, apply 및 lambda 함수를 써서 나이 계산

```python
data['age']=data.birthDate.apply(lambda x:(today-x).days//365)
data.head()
```

##### 3) age 결측치를 선수 평균 나이로 대체

```python
data['age']=data['age'].fillna(data.age.mean())
data
```

#### height 컬럼(feet-inch 형식)으로 heightCm 컬럼 추가하기

##### 1) heightCm 컬럼 생성, apply 및 lambda 함수를 사용해서 height 컬럼의 feet과 inch를 cm로 변환 ('-'를 기준으로 split해서 각각을 cm로 변환해 더하기)



```python
#도저히 lambda 함수를 이용하는 방법을 찾지 못했습니다..ㅜㅜ 그래서 다른 방법으로 풀이하였으니 양해 부탁드립니다!
feet=data.height.str.split('-').str[0]
inch=data.height.str.split('-').str[1]
data["feet"]=feet
data["inch"]=inch
data["feet"]=data["feet"].astype(int)
data["inch"]=data["inch"].astype(int)
data['heightCm']=data.feet*30+data.inch*2
data.head()
```
