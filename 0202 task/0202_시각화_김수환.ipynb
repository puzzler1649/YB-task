```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# data set: kaggle <Titanic - Machine Learning from Disaster>
# https://www.kaggle.com/competitions/titanic/overview
df = pd.read_csv('/content/drive/MyDrive/ybigta/task data/visual_train.csv')
df
```

```python
df.info()
```

```python
palette = sns.color_palette("bright")
sns.palplot(palette)
```

```python
sns.set_palette("bright")
sns.countplot(x='Embarked', data= df)
#sns.countplot(x='Sex',data = df)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

plt.rc('font', size=15)

# Bubble Chart
sns.scatterplot(data = df, x='Age', y='Fare', size='Pclass', hue='Embarked')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

sur=df[df['Survived']==1]
plt.rc('font', size=15)

# Bubble Chart
sns.scatterplot(data = sur, x='Age', y='Fare', size='Pclass', hue='Embarked')

plt.show()
```
